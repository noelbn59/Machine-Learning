```python
# MINMAX SCALER



import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("USA_cars_datasets.csv")
print("Original Data:\n", df.head())

scaler = MinMaxScaler()
numeric_df = df.select_dtypes(include=['int64', 'float64'])
scaled_data = scaler.fit_transform(numeric_df)

scaled_df = pd.DataFrame(scaled_data, columns=numeric_df.columns)
print("\nScaled Data (0-1):\n", scaled_df.head())

final_df = pd.concat([df.select_dtypes(exclude=['int64', 'float64']), scaled_df], axis=1)
final_df.to_csv("USA_cars_scaled.csv", index=False)
print("\n✅ Scaled dataset saved as 'USA_cars_scaled.csv'")
```

```python
# Standard Scaler


import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("USA_cars_datasets.csv")
print("Original Data:\n", df.head())

scaler = StandardScaler()
numeric_df = df.select_dtypes(include=['int64', 'float64'])
scaled_data = scaler.fit_transform(numeric_df)

scaled_df = pd.DataFrame(scaled_data, columns=numeric_df.columns)
print("\nStandard Scaled Data:\n", scaled_df.head())

final_df = pd.concat([df.select_dtypes(exclude=['int64', 'float64']), scaled_df], axis=1)
final_df.to_csv("USA_cars_scaled.csv", index=False)
print("\n✅ Scaled dataset saved as 'USA_cars_scaled.csv'")
```

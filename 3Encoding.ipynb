```python
# LABEL ENCODER


import pandas as pd
from sklearn.preprocessing import LabelEncoder

data={'Gender':['M','F','F','M']}

df=pd.DataFrame(data)
print("Orginal Data:")
print(df)

le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])

print("\n After Label Encoding:")
print(df)
```

```python
# ONE-HOT ENCODER



import pandas as pd

data={'Color':['Red','Blue','Green','Red']}

df=pd.DataFrame(data)
print("Orginal Data:")
print(df)


df_encoded=pd.get_dummies(df,columns=['Color'])
print("\n After One-Hot Encoding:")
print(df_encoded)





```

```python
# ONE-HOT ENCODER USING SKIKIT


from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data=pd.DataFrame({'Color':['Red','Blue','Green','Red']})
print("Orginal Data:")
print(data)

encoder=OneHotEncoder()
encoded=encoder.fit_transform(data[['Color']])

encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out(['Color']))
print(encoded_df)
```

```python
# DUMMY ENCODING


import pandas as pd

data={'Color':['Red','Blue','Green','Red']}

df=pd.DataFrame(data)
print("Orginal Data:")
print(df)


df_encoded=pd.get_dummies(df,columns=['Color'],drop_first=True)
print("\n After One-Hot Encoding:")
print(df_encoded)
```

```python
# Standard SCALING

import pandas as pd
from sklearn.preprocessing import StandardScaler


data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [3000, 4000, 5000, 6000, 2000],
    'Exp': [2, 3, 5, 7, 1]
})

print("Original Data:\n", data)

scaler = StandardScaler()

scaled_data = scaler.fit_transform(data)
scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

print("\nScaled Data:\n", scaled_df)
```

```python
# MINMAX SCALING

import pandas as pd
from sklearn.preprocessing import MinMaxScaler


data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [3000, 4000, 5000, 6000, 2000],
    'Exp': [2, 3, 5, 7, 1]
})

print("Original Data:\n", data)

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(data)
scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

print("\nScaled Data(0-1):\n", scaled_df)
```
